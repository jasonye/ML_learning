```python
import torch
```

```python
# 创建一个需要计算梯度的张量

# 当你创建一个 requires_grad=True 的张量时，PyTorch 会自动跟踪所有对它的操作，以便在之后计算梯度。
x = torch.randn(2, 2, requires_grad=True)
print(x)

# 执行某些操作
y = x + 2

print(y)

z = y * y * 3
print(z)

out = z.mean()

print(out)

```

```python
# 反向传播，计算梯度
out.backward()

# 查看 x 的梯度
print(x.grad)
```

```python
x1 = torch.randn(2, 2, requires_grad=False)
print(x1)

# 执行某些操作
y1 = x1 + 2

print(y1)

z1 = y1 * y1 * 3
print(z1)

out = z1.mean()

print(out)
```
