```python
# 练习2: 基于鸢尾花瓣特征集的分类

import numpy as np 
import pandas as pd 

data = pd.read_csv("../data/Iris/Iris.csv")
print(data.shape)
print(data.head())
```

```python
X_petal = data[['PetalLengthCm', 'PetalWidthCm']]
print(X_petal.shape)
print(X_petal.head())

y = data["Species"]
print(y.shape)
print(y.head())
print(y[0:10])

from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
y_encoded = ordinal_encoder.fit_transform(data[['Species']])
print(y_encoded.shape)
print(y_encoded[:10])

# 这里导入的数据是文本，需要进行规范化？
```

```python
from sklearn.model_selection import train_test_split

#print(X_petal.head())
# print(y_encoded[:10])

X_train, X_test, y_train, y_test = train_test_split(X_petal, y_encoded, test_size=0.2, random_state=0)

print("训练集长度:", len(X_train))
```

```python
from sklearn.preprocessing import StandardScaler

# 对数据进行与预处理
## 归一化处理，对数据进行标准化
scaler = StandardScaler()
X_train_petal = scaler.fit_transform(X_train)
X_test_petal = scaler.transform(X_test)

print(X_train[:5])
print(X_train_petal[:5])
print(X_test[:5])
print(X_test_petal[:5])
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty="l2", C=0.2)  # Logistic Regression with L2 regularization
lr.fit(X_train_petal, y_train)

y_test_out = lr.predict(X_test_petal)
print(y_test_out[:10])
print(y_test[:10])

print(ordinal_encoder.inverse_transform(y_test_out.reshape(-1, 1)))

score = lr.score(X_test_petal, y_test)
print("LogicRegression {:.2f}".format(score*100))
```

```python
#Ex3: 进行C值比较，分析模型效果

from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))

j = 0 
C_param_range = [0.01, 0.05, 0.08,  0.1, 0.2, 0.3,  0.5, 1, 2, 5, 10, 100]
for C_param in C_param_range:
    lr = LogisticRegression(penalty="l2", C=C_param)
    lr.fit(X_train_petal, y_train)
    y_test_out = lr.predict(X_test_petal)
    acc = accuracy_score(y_test, y_test_out)
    print("C={}: 测试集准确率: {:.2f}%".format(C_param, acc*100))

    j += 1
    plt.subplot(4, 3, j)
    plt.title("C={}".format(C_param))
    # plt_decision_region(X_train_petal, y_train, classifier = lr, test_idx=range(30, 40))
    plt.scatter(X_test_petal[:, 0], X_test_petal[:, 1], c=y_test_out, marker='o', s=50, edgecolor='k')
```
