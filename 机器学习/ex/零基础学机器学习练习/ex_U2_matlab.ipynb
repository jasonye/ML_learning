```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.linspace(-10, 10, 100)
y = 2*x+1


fig, ax = plt.subplots(figsize=(5, 5))
# 这两个添加横线
ax.axhline(y=0, color="black", linestyle="--")
ax.axvline(color="grey", linestyle="-.")

plt.xlabel("x")
plt.ylabel("y")

# 方法1: 使用set_xlim/set_ylim
ax.set_xlim(-2, 2)  # x轴范围设为0-8
ax.set_ylim(-1.5, 2)  # y轴范围设为-1.5-1.5

plt.plot(x, y, linestyle = '-')
plt.annotate("Intercept", xy=(0, 1)) # 截距
# axes[0].set_yticks([-10, 10, 1])
plt.show()
```
