# 项目1: 使用K-Means算法对数据进行聚类

实验方法：
1.  生成测试数据、观察测试数据
2.  使用Matplotlib把数据显示出来，观察数据分类情况
3.  使用手肘法模拟 求出最佳 集簇数k
4.  最后使用观察到的集簇数据、进行数据分类、并且把绘图展示效果。

```python
from sklearn.datasets import make_blobs

# 
# 参看文档生成测试数据： https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs
X, y = make_blobs(n_samples=100, n_features=2, centers=3, center_box=(-10, 10), cluster_std=1.0)


print(X.shape)
print(X[:10])
print(y)
```

```python
# 把测试数据使用pyplot显示出来

import matplotlib.pyplot as plt
import numpy as np

col1_zip, col2_zip = zip(*X)  # 注意：返回的是元组
colx = list(col1_zip)  # 转换为列表
coly = list(col2_zip)
print(colx[:5])
print(coly[:5])

plt.scatter(colx, coly)
plt.show()
```

```python
from sklearn.cluster import KMeans

# 使用手肘法模拟 求出最佳 集簇数

inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# 使用K-Means方法拟合，并且把结果显示出来

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)


plt.scatter(colx, coly, c=kmeans.labels_)
plt.show()
```
