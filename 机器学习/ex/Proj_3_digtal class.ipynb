```python
from sklearn.datasets import fetch_openml

# 从openml下载可用的测试数据
mnist = fetch_openml("mnist_784", as_frame=False)

X, y = mnist.data, mnist.target 
```

```python
X.shape
```

```python
y.shape
```

```python
y[0]
```

```python
import matplotlib.pyplot as plt 

# 做数据的显示
def plot_digit(image_data):
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap="binary")
    plt.axis("off")
```

```python
some_digit = X[0]
plot_digit(some_digit)
plt.show()
```

```python
X_train, y_train, X_test, y_test = X[:60000], y[0:60000], X[60000:], y[60000:]
```

```python
y_train.shape
```

```python

# 准备训练数据
y_train_5 = (y_train == '5')

y_test_5 = (y_test == '5')

y_train_5
```

```python
y_test_5
```

```python
from sklearn.linear_model import SGDClassifier

# 进行模型训练
sgd_clf = SGDClassifier(random_state = 42)

sgd_clf.fit(X_train, y_train_5)
```

```python
# 用模型进行数据预测
sgd_clf.predict([some_digit])
```

```python
some_digit = X[2]

sgd_clf.predict([some_digit])
```

```python
plot_digit(some_digit)
plt.show()
```

```python
# 练习使用保存模型数据，并且在后续继续使用


import joblib

file = 'models\\sgd_clf_model.joblib'
joblib.dump(sgd_clf, file)
```

```python
model2 = joblib.load(file)

model2.predict([some_digit])

some_digit = X[0]
model2.predict([some_digit])
```
