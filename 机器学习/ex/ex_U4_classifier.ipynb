```python
import numpy as np 
import pandas as pd 

data = pd.read_csv("./data/Iris/Iris.csv")
print(data.shape)
print(data.head())
```

```python
X_petal = data[['PetalLengthCm', 'PetalWidthCm']]
print(X_petal.shape)
print(X_petal.head())

y = data["Species"]
print(y.shape)
print(y.head())

# 这里导入的数据是文本，需要进行规范化？
```

```python
from sklearn.model_selection import train_test_split

X_train, y_train, X_test, y_test = train_test_split(X_petal, y, test_size=0.2, random_state=0)

print("训练集长度:", len(X_train))
```

```python
from sklearn.preprocessing import StandardScaler

# 对数据进行与预处理
scaler = StandardScaler()
X_train_petal = scaler.fit_transform(X_train)
X_test_petal = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty="l2", C=0.1) # 
lr.fit(X_train_petal, y)

lr.predict(X_test_petal)

score = lr.score(X_test_petal, y_test)
print("LogicRegression {:.2f}".format(score*100))
```
